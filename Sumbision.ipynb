Kemudian 1 kolom memilikitipe data objec

Pada kolom dengan tipe data float dan integer, missing value diatasi dengan mengisi nilai mean dari setiap kolomnya. Hal ini dilakukan karena data yang digunakan merupakan data continyu dan terdapat missing value dengan jumlah yang cukup besar. Kemudian untuk kolom yang memiliki tipe data object missing value diatasi dengan mengisi dengan nilai modus pada kolom tersebut.

Pada tahap ini akan dilakukan pengelompokan nilai PM2.5 dan PM10. Berdasarkan WHO ambang batas dari PM2.5 adalah 10 mikrogram per meter kubik (g/m) dan PM10 25 g/m dalam periode waktu 24 jam. Pengambilan controh hanya dilakukan pada satu data saja karena data yang digunakan memiliki pola yang sama.

```python
#mencari nilai pm2.5 maksimal setiap tahun untuk seluruh kota
pm25_all_max = pd.DataFrame(columns=aotizhongxin_df.columns)

for year in year_list:
    for station in station_list:
        year_pm25 = station[station['year'] == year]
        max_pm25 = year_pm25[year_pm25['PM2.5'] == year_pm25['PM2.5'].max()]
        pm25_all_max= pd.merge(max_pm25, pm25_all_max, how='outer')
    
pm25_all_max
```

```python
#mengelompokan nilai pm2.5 rata - rata berdasarkan station, year, month 
pm25_max = pm25_all_max.groupby(['station','year','month'])['PM2.5'].mean()
```

```python
#mencari nilai pm2.5 minimal setiap tahun untuk seluruh kota
pm25_all_min = pd.DataFrame(columns=aotizhongxin_df.columns)

for year in year_list:
    for station in station_list:
        year_pm25 = station[station['year'] == year]
        min_pm25 = year_pm25[year_pm25['PM2.5'] == year_pm25['PM2.5'].min()]
        pm25_all_min = pd.merge(min_pm25, pm25_all_min, how='outer')
    
pm25_all_min
```

```python
#mengelompokan nilai pm2.5 rata - rata berdasarkan station, year, month 
pm25_min = pm25_all_min.groupby(['station','year','month'])['PM2.5'].mean()
```
